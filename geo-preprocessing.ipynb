```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import box
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
file_path = '/Users/nf/docs/uni/studies/master/5th/2-GIScm/assignments/mapping-project/data/UNHCR-PRMN-Displacement-Dataset.xlsx'
```

```python
data = pd.read_excel(file_path)
```

```python
import matplotlib.pyplot as plt

# Filter data for the year 2023
data_2023 = data[data['Month End'].dt.year == 2023]
flood_2023 = data_2023[data_2023['Reason'] == 'Flood']

# Group by 'Previous (Departure) Region' and sum the 'Number of Individuals' in thousands
departure_region_totals = flood_2023.groupby('Previous (Departure) Region')['Number of Individuals'].sum() / 1000

# Plot the bar graph
plt.figure(figsize=(12, 6))
departure_region_totals.sort_values(ascending=False).plot(kind='bar')
plt.title('Flood-Displaced People in Somalia by Departure Region (2023)', fontsize=16)
plt.xlabel('Departure Region', fontsize=14)
plt.ylabel('Total Number of Individuals (Thousands)', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()
plt.show()

```
